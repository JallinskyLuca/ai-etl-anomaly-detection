### üìà Distribution of Engineered Features: `time_since_last`

```python
# Only values in range -50 to 500
filtered = df['time_since_last'][(df['time_since_last'] >= -50) & (df['time_since_last'] <= 500)]

plt.figure(figsize=(12,6))

# Explicit bins for clarity
bins = range(-50, 501, 10)  # 10-second bins

sns.histplot(filtered, bins=bins, color='skyblue', edgecolor='black', alpha=0.7)

plt.yscale('log')  # optional: make small counts visible
plt.title("Time Since Last Transaction ‚Äî Bursts Only (Log Scale)")
plt.xlabel("Seconds since last transaction")
plt.ylabel("Count (log scale)")
plt.grid(True, which="both", ls="--", alpha=0.5)

plt.show()
```

### üßÆ Standardization Formula (Math Cell)

We use Z-score scaling:

$$
x' = \frac{x - \mu}{\sigma}
$$

Where:

- \( x \) is the original value
- \( \mu \) is the feature mean
- \( \sigma \) is the standard deviation
